```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# focused image from focused.png
focused = cv2.imread('focused.png')

# unfocused image from unfocused.png
unfocused = cv2.imread('unfocused.png')

# convert to grayscale
focused_gray = cv2.cvtColor(focused, cv2.COLOR_BGR2GRAY)

unfocused_gray = cv2.cvtColor(unfocused, cv2.COLOR_BGR2GRAY)

# compute the Laplacian of the image
laplacian_focused = cv2.Laplacian(focused_gray, cv2.CV_64F)

laplacian_unfocused = cv2.Laplacian(unfocused_gray, cv2.CV_64F)

# Plot laplacian
plt.figure()
plt.subplot(1,2,1)
plt.imshow(laplacian_focused, cmap='gray')

# Label with score, variance of laplacian
score_focused = np.var(laplacian_focused)
score_unfocused = np.var(laplacian_unfocused)

plt.title('Focused variance: {:.2f}'.format(score_focused))

plt.subplot(1,2,2)
plt.imshow(laplacian_unfocused, cmap='gray')

plt.title('Unfocused variance: {:.2f}'.format(score_unfocused))


plt.show()

# Show cross-section plot at y=300

plt.figure()
plt.plot(laplacian_focused[300,:], label='focused')
plt.plot(laplacian_unfocused[300,:], label='unfocused')
# title
plt.title('Cross-section plot of laplacian at y=300')
plt.legend()

# Crosssection plot at y=300, original image
plt.figure()
plt.plot(focused_gray[300,:], label='focused')
plt.plot(unfocused_gray[300,:], label='unfocused')
# title
plt.title('Cross-section plot of image at y=300')
plt.legend()

# axis labels
plt.xlabel('x')
plt.ylabel('intensity')

# Crosssection plot at x=300, 2nd derivative
plt.figure()
deriv_focused = np.gradient(focused_gray[300,:])
deriv_unfocused = np.gradient(unfocused_gray[300,:])
deriv_focused2 = np.gradient(deriv_focused)
deriv_unfocused2 = np.gradient(deriv_unfocused)

plt.plot(deriv_focused2, label='focused')
plt.plot(deriv_unfocused2, label='unfocused')
# title
plt.title('Cross-section plot of 2nd derivative at y=300')
plt.legend()

# axis labels
plt.xlabel('x')
plt.ylabel('intensity\'\'')


```
